```python
import pandas as pd
```

```python
ct = pd.read_csv("data/ct-1960-2014.csv",skiprows=8).dropna(thresh=10)
ct["location"] = "Connecticut"
```

```python
us = pd.read_csv("data/us-1960-2014.csv",skiprows=8).dropna(thresh=10)
us["location"] = "United States"
us
```

```python
recent_df = pd.read_excel("data/UCR by state 2005-2017.xlsx")
recent_df = recent_df[recent_df["Area"].apply(lambda x: x in ["Connecticut",
                                                             "United States"])]
recent_df
```

```python
for c in recent_df.columns:
    print c
```

```python
def get_recent(area,topic):
    df = pd.pivot_table(recent_df,
                       columns="Year",
                       index="Area").loc[area,topic].to_frame()
    return df
get_recent("Connecticut","Violent crime rate")
```

```python
def timeline(df, location, topic="Violent Crime rate", recent_col=None, ):
    
    tmp_df = df.copy()

    tmp_df[topic] = pd.to_numeric(tmp_df[topic])
    
    ret =  pd.pivot_table(tmp_df,
                         columns="Year",
                         values=topic,
                         index="location").transpose()

    if recent_col is not None:
        recent = get_recent(location, recent_col)        
        ret = pd.concat([ret,
                        recent[recent.index > 2014]])
        
    return ret

timeline(ct, "Connecticut", topic="Violent Crime rate", recent_col="Violent crime rate")
```

```python
timeline(us)
```

```python
def both_timelines(topic="Violent Crime rate", recent_col=None):
    
    ctdf = timeline(ct, "Connecticut", topic=topic, recent_col=recent_col).transpose()
    usdf = timeline(us, "United States", topic=topic, recent_col=recent_col).transpose()


    ret = pd.concat([ctdf, usdf])
    return ret

both_timelines(topic="Violent Crime rate", recent_col="Violent crime rate").transpose().plot()
```

```python
print both_timelines(topic="Violent Crime rate", recent_col="Violent crime rate").transpose().to_csv(sep='\t')
```
